# ETL YouTube Trending Videos Project

```python
# Importing dependencies 
import json
import pandas as pd
from sqlalchemy import create_engine
```

## Extracting CSVs and JSONs into Pandas Dataframes

###### Mexico

```python
# Reading Mexico csv
mx_file = "Resources/MXvideos.csv"
mx_df = pd.read_csv(mx_file, encoding = "utf8")

# Adding country name column
mx_df["country_name"] = "Mexico"

# Printing dataframe
mx_df.head()
```

```python
# Reading Mexico json
with open("Resources/MX_category_id.json") as x:
    data = json.load(x)

# Creating empty lists
mx_id_list = []
mx_title_list = []

# Extracting wanted info and appending it to empty lists
for item in data["items"]:
    mx_id_cat = int(item["id"])
    mx_cat_title = item["snippet"]["title"]
    
    mx_id_list.append(mx_id_cat)
    mx_title_list.append(mx_cat_title)
```

```python
# Generating dataframe with the lists extracted from the json file
mx_categories_df = pd.DataFrame({"category_id": mx_id_list, 
                                 "category_name": mx_title_list})

# Printing dataframe
mx_categories_df
```

###### United States

```python
# Reading US csv
us_file = "Resources/USvideos.csv"
us_df = pd.read_csv(us_file)

# Adding country name column
us_df['country_name'] = 'United States'

# Printing dataframe
us_df.head(3)
```

```python
# Reading US json file
with open("Resources/US_category_id.json") as x:
    data = json.load(x)

# Creating empty lists
us_id_list = []
us_title_list = []

# Extracting wanted info and appending it to empty lists
for item in data["items"]:
    us_id_cat = int(item["id"])
    us_cat_title = item["snippet"]["title"]
    
    us_id_list.append(us_id_cat)
    us_title_list.append(us_cat_title)
```

```python
# Creating US categories dataframe
us_categories_df = pd.DataFrame({"category_id": us_id_list, 
                                 "category_name": us_title_list})

# Printing dataframe
us_categories_df
```

###### Canada

```python
# Reading Canada csv
ca_file = "Resources/CAvideos.csv"
ca_df = pd.read_csv(ca_file)

# Adding country name column
ca_df['country_name'] = 'Canada'

# Printing dataframe
ca_df.head(3)
```

```python
with open("Resources/CA_category_id.json") as x:
    data = json.load(x)

# Creating empty lists
ca_id_list = []
ca_title_list = []

# Extracting wanted info and appending it to empty lists
for item in data["items"]:
    ca_id_cat = int(item["id"])
    ca_cat_title = item["snippet"]["title"]
    
    ca_id_list.append(ca_id_cat)
    ca_title_list.append(ca_cat_title)
```

```python
# Creating Canada categories dataframe
ca_categories_df = pd.DataFrame({"category_id": ca_id_list, 
                                 "category_name": ca_title_list})

# Printing dataframe
ca_categories_df
```

## Transforming Dataframes

###### Channel Catalog

```python
# Combining all data from the 3 different dataframes 
append_mx_us = mx_df.append(us_df)
append_mx_us_ca = append_mx_us.append(ca_df)

append_mx_us_ca.shape
```

```python
# Dropping channel title duplicates to generate our channel catalog
dropped_series = append_mx_us_ca["channel_title"].drop_duplicates()
dropped_series.shape
```

```python
# Creating channel catalog and adding an index column
channel_catalog = pd.DataFrame(dropped_series).reset_index(drop = True)
channel_catalog["channel_id"] = channel_catalog.index

# Printing dataframe
channel_catalog
```

###### Trending Videos

```python
# Creating video table
videos_df = append_mx_us_ca[["video_id", 
                             "title", 
                             "channel_title", 
                             "category_id", 
                             "publish_time", 
                             "description", 
                             "thumbnail_link"]]

# Printing dataframe
videos_df.head()
```

```python
# Merging trending_videos table with channel_catalog in order to add channel_id and dropping channel_title
trending_videos = pd.merge(videos_df, channel_catalog, how = "inner", on = "channel_title")
trending_videos = trending_videos.drop(["channel_title"], axis = 1)
trending_videos = trending_videos.drop_duplicates(["video_id"])

# Printing dataframe
trending_videos.head()
```

###### Video Analytics

```python
# Creating video analytics table
video_analytics = append_mx_us_ca[["video_id", 
                                   "views",
                                   "likes", 
                                   "dislikes", 
                                   "comment_count",
                                   "comments_disabled", 
                                   "ratings_disabled",
                                   "video_error_or_removed", 
                                   "country_name"]]

# Printing dataframe
video_analytics.head()
```

###### Category Catalog

```python
#  Verifying categories are the same across the three countries
append_1 = mx_categories_df.append(us_categories_df)
append_2 = append_1.append(ca_categories_df)
category_catalog = append_2.drop_duplicates(subset = "category_name").sort_values(by = "category_id", ascending = True)

# Printing dataframe
category_catalog
```

## Creating Database Connection

```python
connection_string = "postgres:Liev1912@localhost:5432/youtube_db"
engine = create_engine(f'postgresql://{connection_string}')
```

```python
engine.table_names()
```

## Loading Dataframes into the database

```python
channel_catalog.to_sql(name = "channel_catalog", con = engine, if_exists = "append", index = False)
```

```python
category_catalog.to_sql(name = "category_catalog", con = engine, if_exists = "append", index = False)
```

```python
trending_videos.to_sql(name = "trending_videos", con = engine, if_exists = "append", index = False)
```

```python
video_analytics.to_sql(name = "video_analytics", con = engine, if_exists = "append", index = False)
```
