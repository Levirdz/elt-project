```python
import os, json
import pandas as pd
from sqlalchemy import create_engine
```

##### Extracting CSVs into Pandas Data Frames

```python
# Reading Mexico csv
mx_file = "Resources/MXvideos.csv"
mx_df = pd.read_csv(mx_file, encoding = "utf8")

# Adding country name column
mx_df['country_name'] = 'Mexico'

# Printing dataframe
mx_df.head(3)
```

```python
# Reading US csv
us_file = "Resources/USvideos.csv"
us_df = pd.read_csv(us_file)

# Adding country name column
us_df['country_name'] = 'United States'

# Printing dataframe
us_df.head(3)
```

```python
# Reading Canada csv
ca_file = "Resources/CAvideos.csv"
ca_df = pd.read_csv(ca_file)

# Adding country name column
ca_df['country_name'] = 'Canada'

# Printing dataframe
ca_df.head(3)
```

```python
append_1 = mx_df.append(us_df)
append_2 = append_1.append(ca_df)

append_2.shape
```

```python
dropped_series = append_2["channel_title"].drop_duplicates()
dropped_series.shape
```

```python
# Creating channel catalog
channel_catalog = pd.DataFrame(dropped_series).reset_index(drop = True)
channel_catalog['channel_id'] = channel_catalog.index
channel_catalog.head(3)
```

```python
# Creating video table
video_table = append_2[["video_id", 
                        "title", 
                        "channel_title", 
                        "category_id", 
                        "publish_time", 
                        "description", 
                        "thumbnail_link"]]
video_table
```

```python
# Merging video table with channel catalog in order to add channel_id and dropping channel_title
video_plus_channel = pd.merge(video_table, channel_catalog, how = "inner", on = "channel_title")
video_plus_channel = video_plus_channel.drop(["channel_title"], axis = 1)
video_plus_channel = video_plus_channel.drop_duplicates(["video_id"])

video_plus_channel.head(3)
```

```python
# Creating video analytics table
video_analytics_table = append_2[["video_id", 
                                  "views",
                                  'likes', 
                                  'dislikes', 
                                  'comment_count',
                                  'comments_disabled', 
                                  'ratings_disabled',
                                  "video_error_or_removed", 
                                  "country_name"]]

video_analytics_table.head(3)
```

##### Reading JSON files

```python
# Reading US json file
with open("Resources/US_category_id.json") as x:
    data = json.load(x)

# Creating empty lists
id_list = []
title_list = []

for item in data["items"]:
    id_cat = int(item["id"])
    cat_title = item["snippet"]["title"]
    
    id_list.append(id_cat)
    title_list.append(cat_title)
```

```python
# Creating US categories dataframe
us_categories_df = pd.DataFrame({"category_id": id_list, 
                             "category_name": title_list})

us_categories_df
```

```python
with open("Resources/CA_category_id.json") as x:
    data = json.load(x)

# Creating empty lists
id_list = []
title_list = []

for item in data["items"]:
    id_cat = int(item["id"])
    cat_title = item["snippet"]["title"]
    
    id_list.append(id_cat)
    title_list.append(cat_title)
```

```python
ca_categories_df = pd.DataFrame({"category_id": id_list, 
                             "category_name": title_list})
```

```python
with open("Resources/MX_category_id.json") as x:
    data = json.load(x)

# Creating empty lists
id_list = []
title_list = []

for item in data["items"]:
    id_cat = int(item["id"])
    cat_title = item["snippet"]["title"]
    
    id_list.append(id_cat)
    title_list.append(cat_title)
```

```python
mx_categories_df = pd.DataFrame({"category_id": id_list, 
                             "category_name": title_list})

mx_categories_df
```

```python
#  Verifying categories are the same across the three countries
append_3 = mx_categories_df.append(us_categories_df)
append_4 = append_3.append(ca_categories_df)
final_json = append_4.drop_duplicates(subset = "category_name").sort_values(by = "category_id", ascending = True)
final_json.shape
```

##### Creating Database Connection

```python
connection_string = "postgres:Liev1912@localhost:5432/youtube_db"
engine = create_engine(f'postgresql://{connection_string}')
```

```python
engine.table_names()
```

##### Loading Dataframes into the database

```python
channel_catalog.to_sql(name = "Channel_catalog", con = engine, if_exists = "append", index = False)
```

```python
final_json.to_sql(name = "Category_catalog", con = engine, if_exists = "append", index = False)
```

```python
video_plus_channel.to_sql(name = "Trending_videos", con = engine, if_exists = "append", index = False)
```

```python
video_analytics_table.to_sql(name = "Analytics", con = engine, if_exists = "append", index = False)
```
